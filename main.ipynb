# Feature Engineering

* create a new columns hours of days 

```python
df.head(1)
```

```python
df['hours'] = df['timestamp'].dt.hour
```

```python
df['day_name'] = df['timestamp'].dt.day_name()
```

```python
df['month_name'] = df['timestamp'].dt.month_name()
```

```python
df['month'] = df['timestamp'].dt.month
```

```python
df['season'] = df['month'].map(
    {
    1: "Winter",
    2: "Winter",
    3: "Spring",
    4: "Spring",
    5: "Spring",
    6: "Summer",
    7: "Summer",
    8: "Summer",
    9: "Fall",
    10: "Fall",
    11: "Fall",
    12: "Winter"
}
)
```

```python
df['week_part'] = df['day_name'].map({
    'Monday': 'Weekday',
    'Tuesday': 'Weekday',
    'Wednesday': 'Weekday',
    'Thursday': 'Weekday',
    'Friday': 'Weekday',
    'Saturday': 'Weekend',
    'Sunday': 'Weekend'
})
```

```python
df['gap'] = df['sales_count'] - df['redemption_count']
```

```python
df['day_period'] = df['hours'].map(
    {
    0: "Mid Night",
    1: "Mid Night",
    2: "Mid Night",
    3: "Mid Night",
    4: "Mid Night",

    5: "Early Morning",
    6: "Early Morning",
    7: "Early Morning",
    8: "Early Morning",

    9: "Morning",
    10: "Morning",
    11: "Morning",

    12: "Afternoon",
    13: "Afternoon",
    14: "Afternoon",
    15: "Afternoon",

    16: "Evening",
    17: "Evening",
    18: "Evening",

    19: "Night",
    20: "Night",
    21: "Night",

    22: "Late Night",
    23: "Late Night"
}
)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df[['hours','month']] = df[['hours','month']].astype('int64')
```

```python
df[['day_name','month_name','season','week_part']] = df[['day_name','month_name','season','week_part']].astype('string')
```
